Below are a histogram and boxplot representation of the same data. A pharmacy is keeping a record of the prices of the drugs that it sells, and an administrator wants to know how much the more expensive drugs tend to cost, in the context of the other prices.

Please write a short explanation of the pros and cons of these two representations. Which would you choose? How would you modify the formatting, if at all, to make it more visually interesting, clear, or informative?

```python
import numpy as np
import pandas as pd

np.random.seed(0)
num_data = 100
data = np.exp(np.random.uniform(size = num_data) * 4)
df = pd.DataFrame(data.T, columns = ["data"])
```

```python
df.head()
```

```python
print("The 75th percentile is:", df.quantile(q = 0.75))
df.plot.hist()
```

**Pros of the histogram representation**<br>
The histogram representation of the data shows clearly that most of the price of the drugs sold by the pharmacy is between 0 and 5. 
it also shows that there are very less samples of drugs that cost between 15 - 25 or 30 - 50.The representation is clear and simple to understand . one thing that could be done to make the histogram more clear is to add the frequency value labels on top of each bins. <br>

**Cons of the histogram representations**<br>
Changing bin size can yield to differing representation of the prices of the drugs. Also this representation does not show the individual data points. Also the histogram doesn't provide the clear view of statistics such as mean, median or standart deviation.

```python
df.plot.box()
```

**Pros of the Box plot representation**<br>
The Box plot representation clearly shows that the drug prices are between 5 and 15 with the box and it also shows clearly that very less drugs are sold for more that 30 . it also clearly shows the outliers that are above 35 are very sparse in the dataset. <br>

**Cons of the Box plot representation**<br>
The Box plot representation does not show full shape of the distribution of the drug prices. it does not reveal the subtle variations of the drug price distribution which can be determined by histograms by varying bin sizes.

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("joebeachcapital/risk-factor-prediction-of-chronic-kidney-disease")

print("Path to dataset files:", path)
```

```python
df = pd.read_csv(path + "/ckd-dataset-v2.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Count the number of rows with at least one null value
null_records_count = df.isnull().any(axis=1).sum()
print(f"Number of records with null columns: {null_records_count}")
# Print rows with at least one null value
rows_with_nulls = df[df.isnull().any(axis=1)]
print(rows_with_nulls)
```

```python
# Convert numerical columns to float
numerical_columns = df.columns[df.iloc[2:].applymap(lambda x: x.replace('.', '', 1).isdigit()).all()]
df[numerical_columns] = df[numerical_columns].apply(pd.to_numeric, errors='coerce')
```

```python
df.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select numerical columns
numerical_columns = df.select_dtypes(include=['float64']).columns
swarm_cols = numerical_columns[0:4]
kde_cols = print(numerical_columns[5:8])
violin_cols = print(numerical_columns[9:])


for col in swarm_cols:
    plt.figure(figsize=(3, 3))
    sns.swarmplot(data=df, x=col)
    plt.title(f'Swarm Plot for {col}')
    plt.show()
```

```python
# Define the columns for KDE plot
kde_cols = numerical_columns[5:8]  # Adjust the range as needed

# Plot KDE for each column in kde_cols
for col in kde_cols:
    plt.figure(figsize=(6, 4))
    sns.kdeplot(data=df, x=col, fill=True)
    plt.title(f'KDE Plot for {col}')
    plt.xlabel(col)
    plt.ylabel('Density')
    plt.show()
```

```python
# Define the columns for violin plot
violin_cols = numerical_columns[9:13]  # Adjust the range as needed

# Plot violin plots for each column in violin_cols
for col in violin_cols:
    plt.figure(figsize=(6, 4))
    sns.violinplot(data=df, y=col)
    plt.title(f'Violin Plot for {col}')
    plt.xlabel(col)
    plt.show()
```

```python
df['stage'].value_counts().plot(kind='bar', figsize=(8, 6), color='skyblue')
plt.title('Count of Values in Stage Column')
plt.xlabel('Stage')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate mean, median, and mode for numerical columns
mean_values = df[numerical_columns].mean()
median_values = df[numerical_columns].median()
mode_values = df[numerical_columns].mode().iloc[0]

# Display the results
print("Mean values:\n", mean_values)
print("\nMedian values:\n", median_values)
print("\nMode values:\n", mode_values)
```

```python
df.groupby('age')['affected'].sum().plot(kind='bar', figsize=(10, 6), color='red')
plt.title('Affected per Age group')
plt.xlabel('Age')
plt.ylabel('Count of Affected')
plt.xticks(rotation=45)
plt.show()
```
